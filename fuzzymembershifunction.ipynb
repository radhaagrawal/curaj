```python
#triangular Membership
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.arange(11)
mfx = fuzz.trimf(x, [0, 5, 10])
plt.plot(x,mfx)
```

```python
#Trapezoidel Membership
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.arange(11)
mfx = fuzz.trapmf(x, [0, 5, 10,15])
plt.plot(x,mfx)
```

```python
#Sigmoidal Membership
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.arange(11)
mfx = fuzz.smf(x, 0,10)
plt.plot(x,mfx)
```

```python
#Gaussian membership
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.arange(11)
mfx = fuzz.gaussmf(x, 5,2)
plt.plot(x,mfx)
```

```python
#GBell Membership
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.arange(15)
mfx = fuzz.gbellmf(x, 4,5,8)
plt.plot(x,mfx)
```
